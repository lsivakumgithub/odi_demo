```python
# Define the file path and content
file_path = './example.txt'
content = '''This is an example text file.
It contains multiple lines of text.
You can write any content here!'''

# Open the file in write mode and write content
with open(file_path, 'w') as file:
    file.write(content)

# Check if the file is created and its content
with open(file_path, 'r') as file:
    print(file.read())  # Display the content of the file to verify
```
